# Restoring Corrupted Student Picture to original color and contrast

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the corrupted image
image = cv2.imread("Corrupted Student Picture.png")

# Convert from BGR (OpenCV default) to RGB (for display)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 1: Adjust brightness and contrast
# Formula: new_image = alpha * image + beta
# alpha: contrast (1.0 - 3.0), beta: brightness (0 - 100)
alpha = 1.1   # contrast
beta = 30     # brightness

restored = cv2.convertScaleAbs(image_rgb, alpha=alpha, beta=beta)

# Step 2: Optional color correction using histogram equalization (for more natural look)
img_yuv = cv2.cvtColor(restored, cv2.COLOR_RGB2YUV)
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])  # apply histogram equalization on brightness channel
final_restored = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)

# Display results
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("Corrupted Picture")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(restored)
plt.title("After Brightness & Contrast")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(final_restored)
plt.title("Final Restored Picture")
plt.axis("off")

plt.show()
```

##Contrast (Alpha) and Brightness (Beta) Used

**Alpha (contrast): 1.1**
