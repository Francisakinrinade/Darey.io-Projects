Query 1: Find all patients in the states "FCT Abuja" and "Plateau".

```python
# Assuming 'engine' is available from cell ac9976da

import pandas as pd

# Execute the SQL query and load the results into a pandas DataFrame
query = """
SELECT *
FROM Patients
WHERE State IN ('Fct Abuja', 'Plateau');
"""

# Use the engine created in ac9976da to read the SQL query into a DataFrame
df_results = pd.read_sql(query, engine)

# Display the resulting DataFrame
display(df_results)
```

Query 2: Retrieve the total number of male and female patients.

```python
# Assuming 'engine' is available from cell ac9976da

import pandas as pd

# Execute the SQL query using pandas read_sql
query = """
SELECT Gender, COUNT(*) AS TotalPatients
FROM Patients
GROUP BY Gender;
"""

# Use the engine created in ac9976da to read the SQL query into a DataFrame
df_gender_counts = pd.read_sql(query, engine)

# Display the resulting DataFrame
display(df_gender_counts)
```

Query 3: List doctors and their specialties in states where confirmed appointments exist.

*This query joins the Doctors and Appointments tables on their common DoctorID.

*It then filters for appointments with a 'Confirmed' status and selects the unique doctor names and their specialties.

```python
# Assuming 'engine' is available from cell ac9976da

import pandas as pd

# Execute the SQL query to list doctors and their specialties in states with confirmed appointments
query = """
SELECT DISTINCT d.Name, d.Speciality
FROM Doctors d
JOIN Appointments a ON d.DoctorID = a.DoctorID
WHERE a.Status = 'Confirmed';
"""

# Use the engine created in ac9976da to read the SQL query into a DataFrame
df_confirmed_doctors = pd.read_sql(query, engine)

# Display the resulting DataFrame
display(df_confirmed_doctors)
```
