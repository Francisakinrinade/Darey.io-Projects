### Count vehicles

#### Subtask:
Count the number of detected vehicles in the current frame using the results from the YOLO model.


````python
# Reset video capture to the beginning
cap.set(cv2.CAP_PROP_POS_FRAMES, 0)

# Loop through each frame of the video
vehicle_classes = ['car', 'truck', 'bus', 'motorcycle']
density_over_time = [] # Re-initialize density_over_time

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform vehicle detection on the current frame
    results = model(frame)

    # Count the number of detected vehicles
    vehicle_count = 0
    # Access the detections for the current frame (results[0])
    detections = results[0].boxes

    if detections:
        # Iterate through each detection
        for box in detections:
            # Get the class ID
            class_id = int(box.cls)
            # Get the class name using the model's names dictionary
            class_name = results[0].names[class_id]

            # Check if the detected object is a vehicle
            if class_name in vehicle_classes:
                vehicle_count += 1

    # Store the vehicle count for the current frame
    density_over_time.append(vehicle_count)

# The loop will finish when the video ends.
# The density_over_time list now contains the vehicle count for each frame.
print(f"Finished processing video. Total frames processed: {len(density_over_time)}")
````

## Modify vehicle counting loop

### Subtask:
Update the loop in Step 3 to not only count the total number of vehicles per frame but also store the counts for each specific vehicle type across all frames.


**Reasoning**:
Update the video processing loop to store vehicle counts per type per frame in a new list.



````python
# Reset video capture to the beginning
if 'cap' in locals() and cap.isOpened():
    cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
else:
    print("Video capture is not opened. Cannot process frames for vehicle type counts.")

# Define the classes that represent vehicles (ensure this matches YOLO model output)
# vehicle_classes = ['car', 'truck', 'bus', 'motorcycle'] # This was already defined
# Re-initialize lists before processing all frames
density_over_time = []
all_frames_vehicle_counts = [] # New list to store detailed counts per frame

# Loop through each frame of the video
if 'cap' in locals() and cap.isOpened():
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform vehicle detection on the current frame
        results = model(frame)

        # Initialize dictionary for current frame vehicle counts
        frame_vehicle_counts = {}

        # Count the number of detected vehicles and their types
        vehicle_count = 0
        # Access the detections for the current frame (results[0])
        detections = results[0].boxes

        if detections:
            # Iterate through each detection
            for box in detections:
                # Get the class ID
                class_id = int(box.cls)
                # Get the class name using the model's names dictionary
                class_name = results[0].names[class_id]

                # Check if the detected object is a vehicle
                if class_name in vehicle_classes:
                    vehicle_count += 1
                    # Increment count for the specific vehicle type
                    if class_name in frame_vehicle_counts:
                        frame_vehicle_counts[class_name] += 1
                    else:
                        frame_vehicle_counts[class_name] = 1

        # Store the total vehicle count for the current frame
        density_over_time.append(vehicle_count)
        # Store the detailed vehicle counts for the current frame
        all_frames_vehicle_counts.append(frame_vehicle_counts)


    # The loop will finish when the video ends.
    print(f"Finished processing video. Total frames processed: {len(density_over_time)}")
    print(f"Collected detailed vehicle counts for {len(all_frames_vehicle_counts)} frames.")

else:
    print("Video capture not available or opened. Skipping frame processing.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import collections # Import collections to use Counter

# Aggregate vehicle counts across all frames
# Create a DataFrame from the list of dictionaries
if all_frames_vehicle_counts:
    vehicle_counts_df = pd.DataFrame(all_frames_vehicle_counts)

    # Fill NaN values with 0 (for vehicle types not present in every frame)
    vehicle_counts_df = vehicle_counts_df.fillna(0)

    # Calculate the total count for each vehicle type across all frames
    total_vehicle_type_counts = vehicle_counts_df.sum().astype(int)

    # Plot the distribution of vehicle types
    if not total_vehicle_type_counts.empty:
        plt.figure(figsize=(10, 6))
        total_vehicle_type_counts.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'gold'])
        plt.title("Distribution of Vehicle Types Across All Frames")
        plt.xlabel("Vehicle Type")
        plt.ylabel("Total Count")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print("No vehicle types detected to plot distribution.")

else:
    print("No detailed vehicle count data collected to plot distribution.")
````
